## Referring the algorithm

From both *"Data Mining: Concepts and Techniques"* by Jiawei Han, Jian Pei, and Micheline Kamber, and the article "A Survey of Frequent Itemset Mining Algorithms" by Charu C. Aggarwal and Philip S. Yu, it becomes evident that the selection of association rule mining algorithms is intricately linked to the characteristics of the dataset.

In accordance with the findings presented in these sources:

- **For sparse datasets (many transactions with few items per transaction)**: FP-Growth can be more efficient due to its pattern-based approach and avoidance of candidate generation in later iterations.
- **For dense datasets (many items per transaction)**: Apriori might be preferable as it can leverage the monotonicity property for effective pruning.

## Apriori algorithm
The Apriori algorithm (Agrawal & Srikant, 1994) stands as a foundational pillar in the field of association rule mining, offering an iterative approach to identifying frequent itemsets upon which association rules are built.

**Strengths of the Apriori Algorithm**

- **Simplicity and Understandability**: The Apriori algorithm is renowned for its intuitive and easy-to-understand design.
- **Data Scalability**: While less efficient for massive datasets compared to algorithms like FP-Growth, Apriori remains a viable option for small to medium-sized datasets.



## Fp-Growth Algorithm

The FP-Growth algorithm, introduced by Han, J., Pei, J., & Yin, Y. in their paper *"Mining Frequent Patterns without Candidate Generation,"* offers a novel approach to association rule mining. It employs a compact data structure called the FP-Tree to efficiently mine frequent patterns without the need for candidate generation. This innovative technique has gained widespread adoption due to its ability to handle large-scale datasets and deliver superior performance compared to traditional algorithms like Apriori.


<img src="https://psubnwell.github.io/2018/01/18/apriori-and-fp-growth/toy_dataset.png" width="600"/>

> source: [fp-growth](https://psubnwell.github.io/2018/01/18/apriori-and-fp-growth/toy_dataset.png)

**Strengths of FP-Growth Algorithm**

The FP-Growth algorithm, as outlined by Han et al. (2000), possesses several strengths:

- **Efficiency**: eliminates candidate generation, resulting in significantly reduced computational overhead compared to traditional algorithms like Apriori.
- **Compact Data Representation**: The FP-Tree data structure enables efficient storage and traversal of transactional data, making FP-Growth well-suited for large datasets.
- **Scalability**: can handle massive datasets with millions of transactions and thousands of items, showcasing its scalability and robustness.
- **Parallelization**:can be parallelized for distributed processing across multiple computing nodes, further enhancing its scalability and efficiency.

## Understanding Support, Confidence, and Lift


Three fundamental computations in association rule mining are support, confidence, and lift:

* **Support (S):** Measures the frequency of an itemset within the dataset. It reflects how often a particular combination of items appears together.

    $$ support(A \rightarrow C) = support(A \cup C)  \longrightarrow  \: (range [0,1]) $$

* **Confidence (C):** Indicates the likelihood of the consequent (right-hand side) occurring, given the antecedent (left-hand side). High confidence implies a strong rule.

    $$ confidence(A \rightarrow C) = \frac{support(A \rightarrow C)}{support(A)} \longrightarrow  \: (range [0,1]) $$

* **Lift (L):** Measures the strength of association between the antecedent and consequent, taking into account the overall frequency of the consequent. Lift greater than 1 suggests a positive association.

    $$ lift(A \rightarrow C) = \frac{confidence(A \rightarrow C)}{support(C)} \longrightarrow  \: (range [0, \infty]) $$

# Data Transformation

In association rule mining algorithms, it's common to represent item presence in transactions using binary values (0 or 1). This simplifies calculations and reduces memory usage compared to working with raw item names or IDs.

```python
# Filter rows where Country is "United Kingdom", 
#then group by 'InvoiceNo' and 'Description' and sum up 'Quantity'
itemset = df[df['Country'] == "United Kingdom"]\
            .groupby(['InvoiceNo', 'Description'])["Quantity"]\
            .sum().unstack().reset_index() \
            .set_index("InvoiceNo")


itemset.head()
```

```python
itemset_coded = itemset.applymap(lambda x: 1 if x >= 1 else 0)
```

Association rules primarily aim to discover patterns of co-occurrence between multiple items purchased together. Filtering out invoices with only 1 item eliminates them from the analysis as they don't offer any insights into relationships between different products.

Most association rule mining algorithms utilize a minimum support threshold. This threshold defines the minimum number of times an itemset (group of items) needs to appear in the data to be considered "frequent" and eligible for further analysis as a potential rule. Filtering for invoices with 2+ items ensures that we have a sufficient number of items per transaction to potentially discover frequent itemsets that meet the minimum support requirement.

```python
filtered_itemset = itemset_coded[itemset_coded.sum(axis=1) >= 2]
```

```python
# Create a DataFrame with lift, antecedents, and consequents
lift_table = pd.DataFrame(columns=['Antecedents', 'Consequents', 'Lift'])

# Populate the DataFrame
for i, rule in apriori_rules_test.iterrows():
    lift_table.loc[i] = [list(rule['antecedents'])[0], list(rule['consequents'])[0],rule['lift']]

# Display the DataFrame
lift_table
```

```python
# Create a directed graph
G = nx.DiGraph()

plt.figure(figsize=(9, 10))
# Add nodes for antecedents and consequents
for i, row in lift_table.iterrows():
    G.add_node(row['Antecedents'], color='lightblue')
    G.add_node(row['Consequents'], color='yellow')
    G.add_edge(row['Antecedents'], row['Consequents'], weight=row['Lift'])

    # Draw the graph
pos = nx.spring_layout(G, seed=732)  # Set the seed for reproducibility
nx.draw(G, pos, with_labels=False,node_size=1000, node_color=[color for node, color in nx.get_node_attributes(G, 'color').items()], edge_color='gray', arrowsize=20)

# Add labels for nodes (items) with smaller font size
nx.draw_networkx_labels(G, pos, labels={n: n for n in G.nodes()}, font_size=8)


# Add labels for support and lift below the edges
for edge in G.edges():
    lift = G.edges[(edge[0], edge[1])]['weight']
    x = 0.5 * (pos[edge[0]][0] + pos[edge[1]][0])
    y = 0.5 * (pos[edge[0]][1] + pos[edge[1]][1])
    plt.text(x, y, f"- - - -> {lift:.2f}", fontsize=8, horizontalalignment='center', verticalalignment='bottom', rotation=90)
    
# Add title
plt.title('Association Rules Graph of Apriori Algorithm')

# Show plot
plt.show()
```

This reinforces the notion that when customers consistently buy both item X and item Y together, there's a greater likelihood that their purchase of X will be accompanied by Y.

>- While confidence generally increases with increasing support, **the rate of increase slows** down at higher support levels. This suggests that **rules with very high support are more likely to be confident, but additional support doesn't always lead to significantly higher confidence**. In other words, **the most frequent buying patterns are also likely to be the most reliable indicators of association**, but further increases in purchase frequency may not necessarily translate into proportionally stronger confidence.

>- The data points are not perfectly aligned, indicating that support and confidence are not completely deterministic. This implies that other factors beyond just frequency can influence the strength of associations between items. These factors could include:

 >>   - **Customer preferences**: Individual preferences or specific needs might also play a role. For example, a customer buying a gardening tool might be more likely to also purchase gloves if they have a history of buying these items together, even if this isn't a common pattern overall.

## Visulizing Association rules 
### Networkx

An intriguing approach to analyzing such data involves utilizing network analysis plots to visualize the relationships between each feature and its corresponding compound feature within the dataset. In this case, we employed the NetworkX module to plot the top 100 strongest relations.

To ensure clarity and aesthetics in representing the vast array of data points, we opted for the 'planar_layout' layout. Additionally, to streamline the visualization process, we utilized a lambda function to convert the frozenset objects to strings, eliminating the need to print frozensets with each created node. This enables a more intuitive and visually appealing depiction of the relationships between features and their compound counterparts within the dataset.

```python
apriori_rules['antecedents'] = apriori_rules.antecedents.apply(lambda x: next(iter(x)))
apriori_rules['consequents'] = apriori_rules.consequents.apply(lambda x: next(iter(x)))
```

```python
import networkx as nx
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
plt.style.use('fivethirtyeight')

# Filter the top 100 association rules based on lift
top_100_rules = apriori_rules.sort_values(by='lift', ascending=False).head(100)

# Create the graph
edges = nx.from_pandas_edgelist(top_100_rules, source='antecedents', target='consequents', edge_attr=None)

# Plot the graph
plt.subplots(figsize=(60, 80))
plt.suptitle('Top 100 association rules based on lift', fontsize=50, fontweight='bold')
pos = nx.spring_layout(edges)  # You can choose a layout based on your preference
nx.draw_networkx_nodes(edges, pos, node_size=2000, alpha=0.7, node_color='tomato')
nx.draw_networkx_edges(edges, pos, width=6, alpha=0.2, edge_color='indigo')
nx.draw_networkx_labels(edges, pos, font_size=43, font_weight='bold')
plt.grid()
plt.axis('off')
plt.tight_layout()
plt.show()
```

**Observations:**

> - **Densely connected areas**: The presence of densely connected regions in the graph suggests potential clusters of items that frequently appear together in association rules. These clusters could represent:
    - **Complementary products**: Items that customers often buy together, such as gardening tools and gloves.
    - **Substitute products**: Items that serve similar purposes and might be bought alternatively, depending on factors like price or availability.
    - **Category relationships**: Groups of items belonging to the same category, like different types of vegetables or beverages.
    
> - **Central nodes**: Nodes with a higher number of connections (more edges) might represent frequently purchased items or key items involved in many association rules. These could be:
    - **Staple items**: Products that customers buy regularly, forming the core of their shopping habits.
    - **Versatile items**: Items that can be used in various contexts and might be paired with diverse products.
